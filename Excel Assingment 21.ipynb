Sub EnterName()
    Dim userName As String
    
    ' Prompt the user to enter their name
    userName = InputBox("Please enter your name:", "Enter Name")
    
    ' Check if the user entered a name
    If userName <> "" Then
        ' Enter the name into cell A1 of the active worksheet
        ActiveSheet.Range("A1").Value = userName
        
        ' Display a message box confirming that the name has been entered
        MsgBox "Your name, " & userName & ", has been entered into cell A1.", vbInformation, "Name Entered"
    Else
        ' Display a message box if the user did not enter a name
        MsgBox "No name entered.", vbExclamation, "No Name Entered"
    End If
End Sub


To use this code:

Press ALT + F11 to open the VBA Editor.
Go to Insert > Module to insert a new module.
Copy and paste the above code into the module window.
Close the VBA Editor.
Press ALT + F8, select EnterName, and click Run to execute the macro.
An input box will appear, prompting you to enter your name. After entering your name, click OK.
A message box will appear, confirming that the name has been entered into cell A1.

Userforms in Excel VBA are custom dialog boxes or forms that you can create and use to interact with users. They are used to create interactive interfaces for users to input data, make selections, or perform actions within Excel applications. Userforms can contain various types of controls such as labels, textboxes, listboxes, buttons, etc., allowing for a wide range of user interactions.

To fill a list box using a for loop in a userform, you would typically follow these steps:

Create a Userform: If you haven't already created a userform, you need to insert one. You can do this by going to the VBA Editor (ALT + F11), right-clicking on your VBA project in the Project Explorer, and selecting Insert > UserForm.

Add a List Box: Drag and drop a List Box control from the Toolbox onto your userform. Resize and position the list box as needed.

Write VBA Code: Write VBA code to populate the list box with items using a for loop.

Run the Macro: Run the macro containing the ShowUserform subroutine to display the userform with the populated list box.

Sub CreatePieChart()
    Dim rngData As Range
    Dim chtChart As ChartObject
    Dim i As Integer

    ' Set the data range
    Set rngData = Sheet1.Range("B2:B7")

    ' Create a new chart
    Set chtChart = Sheet1.ChartObjects.Add(Left:=100, Width:=375, Top:=75, Height:=225)

    ' Set chart type
    chtChart.Chart.ChartType = xlPie

    ' Set chart data source
    chtChart.Chart.SetSourceData Source:=rngData

    ' Apply font and title settings
    With chtChart.Chart
        .ChartTitle.Text = "Pie Chart"
        .ChartTitle.Font.Name = "Times New Roman"
        .ChartTitle.Font.Size = 14
        .ChartTitle.Font.Bold = True
    End With

    ' Apply color formatting
    For i = 1 To rngData.Rows.Count
        chtChart.Chart.SeriesCollection(1).Points(i).Interior.Color = RGB(255 - i * 20, i * 30, i * 40)
    Next i
End Sub


This code will create a pie chart on the active sheet using the data in cells B2 to B7. It will format the chart title with the specified font and size, and it will apply custom colors to the pie slices based on the data. Adjust the range B2:B7 to match the actual range of your data.

To protect your workbook using a password in Excel, follow these step-by-step procedures:

Open Excel Workbook: Open the workbook that you want to protect with a password.

Go to File: Click on the "File" tab located in the upper-left corner of the Excel window.

Choose Info: In the menu on the left-hand side, select "Info."

Click on Protect Workbook: Under the "Info" section, click on "Protect Workbook."

Select Encrypt with Password: From the dropdown menu, choose "Encrypt with Password."

Enter Password: A dialog box will appear prompting you to enter a password. Enter your desired password in the text field. Make sure to remember this password because if you forget it, you won't be able to access your workbook.

Confirm Password: After entering the password, you will be asked to confirm it. Re-enter the password in the confirmation dialog box.

Click OK: Once you've confirmed the password, click "OK." Your workbook is now protected with a password.

Save the Workbook: Save the workbook to apply the password protection. You can do this by pressing "Ctrl + S" or by clicking on the "Save" icon in the Quick Access Toolbar.

Close the Workbook: Close the workbook to ensure that the password protection is applied.

Now, whenever you try to open the workbook, Excel will prompt you to enter the password. Only those who know the password will be able to access the workbook.
