# 3.

a)
Null hypothesis (H0): The percentage of citizens in city ABC that owns a vehicle is 60% or less.
Alternate hypothesis (H1): The percentage of citizens in city ABC that owns a vehicle is greater than 60%.

b)
To test the hypothesis, we can conduct a one-sample proportion z-test.

Given:

Sample size (n) = 250
Number of residents owning a vehicle (x) = 170
We first calculate the sample proportion (p-hat):
    p̂ = x/n = 170/250 = 0.68
    
Next, we need to calculate the standard error of the sample proportion (SEp):
    
    SEp= √ p̂(1-p̂)/n
    
    SEp= √ 0.68(1-0.68)/250
    
    SEp≈ 0.0297
    
    Now, we calculate the z-score:

z= p̂-p/SEp

z = 0.68-0.60/0.0297

z≈ 2.69

Using a standard normal distribution table or calculator, we find the critical z-value for a one-tailed test at 10% significance level (α = 0.10) is approximately 1.28.

Since the calculated z-value (2.69) is greater than the critical z-value (1.28), we reject the null hypothesis.

# 5.

In left-skewed (negatively skewed) data:

The mean is less than the median.
The median is less than the mode.
In right-skewed (positively skewed) data:

The mean is greater than the median.
The median is greater than the mode.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate left-skewed data
left_skewed_data = np.random.gamma(2, 1, 1000) * -1

# Generate right-skewed data
right_skewed_data = np.random.gamma(2, 1, 1000)

# Plotting left-skewed data
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(left_skewed_data, bins=30, color='skyblue')
plt.title('Left-skewed Distribution')

# Plotting right-skewed data
plt.subplot(1, 2, 2)
plt.hist(right_skewed_data, bins=30, color='salmon')
plt.title('Right-skewed Distribution')

plt.tight_layout()
plt.show()
```
